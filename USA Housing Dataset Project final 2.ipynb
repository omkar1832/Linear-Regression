### Problem Statement:
The objective of the USA housing dataset is to predict Price. The datasets of several predictor (independent) variables and one target variable (Price)

### Dataset Information:
**Avg. Area Income**: This column represents the average income of the residents in a particular area.

**Avg. Area House Age**: This column represents the average age of houses in the area.

**Avg. Area Number of Rooms**: This column represents the average number of rooms in the houses in the area.

**Avg. Area Number of Bedrooms**: This column represents the average number of bedrooms in the houses in the area.

**Area Population**: This column represents the population of the area where the houses are located.

**Price**: This column represents the price of the houses in the area. It is the target variable that we aim to predict using the other columns.

**Address**: This column represents the address or location of the houses. It provides information about the specific area or region where the houses are situated.

### 5) Split the dataframe into X & y

```python
# Separate object for target feature 
y = df["Price"]

# Separate object for Predictors 
X = df.drop("Price",axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
print('X Shape =',X.shape,'\ny Shape =',y.shape)
```

### 6) Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=14)
```

```python
X_train.shape,y_train.shape
```

```python
y_test.shape,y_test.shape
```

### 7) Build the Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()      
lr.fit(X_train,y_train)    
y_pred = lr.predict(X_test)
```

### Evaluation Metrics

```python
# Train Score
train_score = lr.score(X_train,y_train)
print('Train Score = ',train_score)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print('R2 Score =',score)
```
